```python
gift_matrix = df.pivot_table(index='client_id', columns='gift_name', values='gift_cost')
gift_matrix.head()
```

```python
df_gift_cost.sort_values('gift_name', ascending=False).head(10)
```

```python
food_gift = gift_matrix['Food Basket']
meal_gift = gift_matrix['Meal Coupons']
```

```python
food_gift.head()
```

```python
meal_gift.head()
```

```python
similar_to_food_gift=gift_matrix.corrwith(food_gift)
similar_to_food_gift.head()
```

```python
similar_to_meal_gift=gift_matrix.corrwith(meal_gift)
similar_to_meal_gift.head()
```

```python
corr_food = pd.DataFrame(similar_to_food_gift, columns=['Correlation'])
corr_food.dropna(inplace=True)
corr_food.head()
```

```python
corr_meal = pd.DataFrame(similar_to_meal_gift, columns=['Correlation'])
corr_meal.dropna(inplace=True)
corr_meal.head()
```

```python
corr_food = corr_food.join(df_gift_cost['gift_cost'])
corr_food.head()
```

```python
corr_meal = corr_meal.join(df_gift_cost['gift_cost'])
corr_meal.head()
```

```python
corr_meal[corr_meal['gift_cost']>100].sort_values(by='Correlation', ascending=False).head(10)
```

```python
corr_food[corr_food['gift_cost']>100].sort_values(by='Correlation', ascending=False).head(10)
```
